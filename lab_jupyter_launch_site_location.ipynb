```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

# Define the Haversine function
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points on the Earth.
    Parameters are in decimal degrees.
    """
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Radius of Earth in kilometers (mean radius)
    radius = 6371.0
    distance = radius * c
    return distance

# Step 1: Define proximity points (e.g., nearby cities or landmarks)
proximities = {
    "Nearest Airport": [28.611667, -81.197222],  # Orlando International Airport
    "Nearest City": [28.538336, -81.379234],     # Orlando, FL
    "Another Landmark": [34.05, -118.25]         # Example Landmark: Los Angeles, CA
}

# Step 2: Calculate distances from each launch site to proximity points
launch_sites = {
    "CCAFS LC-40": [28.562302, -80.577356],
    "CCAFS SLC-40": [28.563197, -80.576820],
    "KSC LC-39A": [28.573255, -80.646895],
    "VAFB SLC-4E": [34.632834, -120.610745]
}

# Store the results
distances = []

for site, (lat1, lon1) in launch_sites.items():
    for prox, (lat2, lon2) in proximities.items():
        distance = haversine(lat1, lon1, lat2, lon2)
        distances.append((site, prox, distance))

# Step 3: Display results
for site, prox, distance in distances:
    print(f"Distance from {site} to {prox}: {distance:.2f} km")
```

```python
from math import radians, sin, cos, sqrt, atan2

# Define a function to calculate haversine distance
def calculate_distance(lat1, lon1, lat2, lon2):
    # Radius of Earth in kilometers
    R = 6371.0
    
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Differences in coordinates
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Distance in kilometers
    distance = R * c
    return distance

# Launch site coordinates (example: CCAFS LC-40)
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Closest coastline coordinates
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance to coastline: {distance_coastline:.2f} km")
```
